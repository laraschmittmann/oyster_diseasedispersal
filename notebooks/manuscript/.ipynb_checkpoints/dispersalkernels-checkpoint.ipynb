### Plot dispersal kernels 00-07 at different depths

```python
dist_kernels_5 = pd.read_csv("../../output_data/040.2_connectivity_analysis_05m/040.2_dist_kernels_05_00-07days.csv")
```

```python
dist_kernels_10 = pd.read_csv("../../output_data/040.2_connectivity_analysis_10m/040.2_dist_kernels_05_00-07days.csv")
```

```python
dist_kernels_15 = pd.read_csv("../../output_data/040.2_connectivity_analysis_15m/040.2_dist_kernels_05_00-07days.csv")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# List of datasets and colors for easier plotting
datasets = [dist_kernels_5, dist_kernels_10, dist_kernels_15]
labels = ['5m', '10m', '15m']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Adjust colors as needed

plt.rcParams.update({
    'font.size': 18,       # Font size for text
    'font.family': 'sans-serif',  # Font family for consistency
    'axes.titlesize': 18,   # Font size for titles
    'axes.labelsize': 18,   # Font size for axis labels
    'xtick.labelsize': 18,  # Font size for tick labels
    'ytick.labelsize': 18   # Font size for tick labels
})

fig, ax = plt.subplots(figsize=(8, 8))

for dataset, label, color in zip(datasets, labels, colors):
    # Calculate median and 95% interval bounds using percentiles
    median_curve = dataset.groupby('dist')['obs_per_origin_area'].median()
    lower_bound = dataset.groupby('dist')['obs_per_origin_area'].quantile(0.025)  # 2.5th percentile
    upper_bound = dataset.groupby('dist')['obs_per_origin_area'].quantile(0.975)  # 97.5th percentile
    
    # Plot the median curve as a solid line
    ax.plot(median_curve.index, median_curve.values, color=color, linewidth=4, alpha = 1, label=f'{label}')

    # Shade the area between the 2.5th and 97.5th percentiles (95% interval)
    ax.fill_between(median_curve.index, lower_bound, upper_bound, color=color, alpha=0.1)

    ax.plot(lower_bound.index, lower_bound.values, color=color, linestyle='-', linewidth=1.5, alpha=0.4)
    ax.plot(upper_bound.index, upper_bound.values, color=color, linestyle='-', linewidth=1.5, alpha=0.4)


# Set axis limits
ax.set_xlim(0, 300)
ax.set_ylim(0, 1)

# Add labels and legend
ax.set_ylabel('Cumulative distribution', fontsize=18)
ax.set_xlabel('Distance (km)', fontsize=18)
legend_elements = [Line2D([0], [0], color=color, lw=10, label=label) for label, color in zip(labels, colors)]
ax.legend(handles=legend_elements, loc='best')

# Save the figure
plt.tight_layout()
fig.savefig("plots/dispersal_kernels_95percent_00-07days.pdf", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# List of datasets and colors for easier plotting
datasets = [dist_kernels_5, dist_kernels_10, dist_kernels_15]
labels = ['5m', '10m', '15m']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Adjust colors as needed

plt.rcParams.update({
    'font.size': 18,       # Font size for text
    'font.family': 'sans-serif',  # Font family for consistency
    'axes.titlesize': 18,   # Font size for titles
    'axes.labelsize': 18,   # Font size for axis labels
    'xtick.labelsize': 18,  # Font size for tick labels
    'ytick.labelsize': 18   # Font size for tick labels
})

fig, ax = plt.subplots(figsize=(12, 10))

for dataset, label, color in zip(datasets, labels, colors):
    # Calculate median and standard deviation for each distance
    median_curve = dataset.groupby('dist')['obs_per_origin_area'].median()
    std_curve = dataset.groupby('dist')['obs_per_origin_area'].std()  # Actual standard deviation
    
    # Calculate the bounds using standard deviation (±1 std dev from median)
    lower_bound = median_curve - std_curve
    upper_bound = median_curve + std_curve
    
    # Clip the lower bound to 0 (to avoid going below 0)
    lower_bound = lower_bound.clip(lower=0)
    
    # Plot the median curve as a solid line
    ax.plot(median_curve.index, median_curve.values, color=color, linewidth=4, label=f'{label}')
    
    # Fill the area between the bounds (±1 standard deviation)
    ax.fill_between(median_curve.index, lower_bound, upper_bound, color=color, alpha=0.1)
    
    # Optionally, you can plot the bounds as dotted lines
    ax.plot(lower_bound.index, lower_bound.values, color=color, linestyle='-', linewidth=1.5, alpha=0.4)
    ax.plot(upper_bound.index, upper_bound.values, color=color, linestyle='-', linewidth=1.5, alpha=0.4)

# Set axis limits
ax.set_xlim(0, 300)
ax.set_ylim(0, 1)

# Add labels and legend
ax.set_ylabel('Cumulative distribution', fontsize=18)
ax.set_xlabel('Distance (km)', fontsize=18)
legend_elements = [Line2D([0], [0], color=color, lw=10, label=label) for label, color in zip(labels, colors)]
ax.legend(handles=legend_elements, loc='best')

# Save the figure (optional)
fig.savefig("plots/dispersal_kernels_stddev_00-07days.pdf", dpi=300, bbox_inches='tight')

plt.show()
```
