```python
# parameters
connectivity_file_path = "../../output_data/"
connectivity_file_pattern = "035_connection_count_days-travelled-??_bio_phys_agg.nc"
```

```python
connectivity_file_list = sorted(Path(connectivity_file_path).glob(connectivity_file_pattern))
```

```python
def load_and_clean_connectivity_files(connectivity_file_list, chunks=None):
    # Set default chunks. Note that we don't chunk in the hex1 direction, 
    # because the alignment would have to do a lot of mixing of the original chunks.
    if chunks is None:
        chunks={"hex0": 1000, "month": 1, "year": 1}

    # concatenate after adding days_travelled to each file
    ds = xr.concat(
        [
            xr.open_dataset(_df, chunks=chunks).assign_coords(days_travelled=_dt)
            for _dt, _df in enumerate(connectivity_file_list)
        ],
        dim="days_travelled"
    )

    #Fix aligned coords and aux data vars.
    # Note that we average in the days_travelled direction to erase undefined
    # values in some days_travelled steps.
    
    for coord in ds.coords:
        if "days_travelled" in ds[coord].dims and len(ds[coord].dims) > 1:
            ds[coord] = ds[coord].mean("days_travelled").compute()

    for var in ds.data_vars:
        if (var.endswith("_hex0") or var.endswith("_hex1")) and ("days_travelled" in ds[var].dims):
            ds[var] = ds[var].mean("days_travelled").compute()

    return ds
```

```python
ds = load_and_clean_connectivity_files(connectivity_file_list)
```

### Select days 14-28

```python
ds_conn = ds.sel(days_travelled = slice(14,None)).sum("days_travelled")
ds_conn = ds_conn.compute()
ds_conn
```

### Normalization

```python
ds_conn["obs_per_origin_area"] = ds_conn.obs / ds_conn.water_fraction_hex0
ds_conn
```

### Labellig habitable zones

- depth < 90m
- known population
- rest site
- aquaculture
- known disease site
- known healthy site

```python
ds_conn["habitable_hex0"] = (
    (ds_conn.depth_median_hex0 < 90)
    | (ds_conn.pop_count_hex0 > 0)
    | (ds_conn.his_count_hex0 > 0)
    | (ds_conn.rst_count_hex0 > 0)
    | (ds_conn.aqc_count_hex0 > 0)
    | (ds_conn.dss_count_hex0 > 0)
    | (ds_conn.hly_count_hex0 > 0)
).rename("habitable_hex0")

ds_conn["habitable_hex1"] = (
    (ds_conn.depth_median_hex1 < 90)
    | (ds_conn.pop_count_hex1 > 0)
    | (ds_conn.his_count_hex1 > 0)
    | (ds_conn.rst_count_hex1 > 0)
    | (ds_conn.aqc_count_hex1 > 0)
    | (ds_conn.dss_count_hex1 > 0)
    | (ds_conn.hly_count_hex1 > 0)
).rename("habitable_hex1")
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from matplotlib.collections import PatchCollection
from matplotlib.colors import ListedColormap, BoundaryNorm
import numpy as np
import pandas as pd
from shapely.geometry import Polygon

# Create a mapping of hex IDs to patches
hex_to_patch = {
    hex_id: patch for hex_id, patch in zip(ds_conn.hex1.values, hex_patches)
}

# Extract category data and ensure correct order
hex_ids = dssaqchis_to_habitable_stats.index
category_data = dssaqchis_to_habitable_stats["rel_pathogen_concentration_category"]

# Prepare the patches and data for plotting
hex_patches_filtered = []
category_values = []

for hex_id in hex_ids:
    if hex_id in hex_to_patch:
        hex_patches_filtered.append(hex_to_patch[hex_id])
        category_values.append(category_data.loc[hex_id])

# Define categorical colors for each category
categories = [1, 2, 3, 4, 5]
category_colors = ['#c8e6c9', '#e6ee9c' ,'#fff176','#ff8a65','#f44336']  # Light green, light yellow, light orange, red
cmap = ListedColormap(category_colors)
norm = BoundaryNorm(boundaries=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5], ncolors=len(categories))

# Create the plot
fig, ax = plt.subplots(
    1, 1,
    subplot_kw={"projection": ccrs.Mercator()},
    figsize=(12, 8),
)

#fig.tight_layout()

# Create the PatchCollection
p = PatchCollection(
    hex_patches_filtered,
    cmap=cmap,
    norm=norm,
    alpha=1.0,  
    linewidth=0.1,
    transform=ccrs.PlateCarree(),
    zorder=0
)

p.set_array(np.array(category_values))

ax.add_collection(p, autolim=True)

plot_hex_ids_with_land_overlap = set(hex_ids).intersection(land_overlap_hex_ids)

land_overlap_hexes = [hex_to_patch[hex_id] for hex_id in plot_hex_ids_with_land_overlap if hex_id in hex_to_patch]

land_overlap = PatchCollection(
    land_overlap_hexes,
    color='grey',
    alpha=1,
    linewidth=0.1,
    transform=ccrs.PlateCarree(),
    zorder=2
)

ax.add_collection(land_overlap, autolim=True)

# Set the extent of the plot (longitude and latitude limits)
ax.set_extent([-12, 9, 45, 59])

# Add coastlines and gridlines
ax.coastlines(zorder=0)
gl = ax.gridlines(draw_labels=True, zorder=0)
gl.xlabel_style = {'size': 16}
gl.ylabel_style = {'size': 16}
gl.top_labels = False   # suppress top labels
gl.right_labels = False # suppress right labels

# Set the title of the plot
#ax.set_title("Pathogen Concentration Categories", fontsize=16)

# Add colorbar with custom formatting
cbar = fig.colorbar(p, ax=ax, orientation='vertical', pad=0.02, shrink=0.5, aspect=15)
cbar.ax.tick_params(labelsize=16)
cbar.set_ticks(categories)  # Set colorbar ticks to match your categories
cbar.set_ticklabels(['None', 'Low', 'Moderate', 'Moderately high', 'High'])  # Label each tick with your category values

# Format colorbar to be more condensed
cbar.ax.invert_yaxis()  # Optional: invert the colorbar to align with the plot orientation

# Save the figure
fig.savefig("plots/042_risk-analysis_dssaqchis.pdf")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define the custom green to red color map with white for the 'None' category
category_colors = ['#ffffff', '#c8e6c9', '#e6ee9c', '#fff176', '#ff8a65', '#f44336']  # White for 0, green to red for categories
cmap = sns.color_palette(category_colors, as_cmap=True)

# Pivot the table to get the heatmap data
heatmap_data_categorical = dssaqchis_to_habitable_stats.pivot_table(
    index='hex1', 
    columns='month_year', 
    values='rel_pathogen_concentration_category',
    aggfunc='first'  # Use 'first' to take the value directly since there is only one
)

# Calculate the mean value of 'rel_pathogen_concentration_category' for each 'hex1'
mean_values = heatmap_data_categorical.mean(axis=1)

# Sort the DataFrame by these mean values in descending order
sorted_hex1 = mean_values.sort_values(ascending=False).index
heatmap_data_categorical = heatmap_data_categorical.loc[sorted_hex1]

# Create the heatmap
plt.figure(figsize=(14, 10))
ax = sns.heatmap(
    heatmap_data_categorical,
    cmap=cmap,
    vmin=0, vmax=5,  # Ensure the colorbar matches the categories
    cbar_kws={'label': '', 'ticks': [0, 1, 2, 3, 4, 5]}  # Adjust ticks to match categories
)

# Set custom labels for the x-axis
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
ax.set_xlabel('')
ax.set_ylabel('Habitable sites')

# Customize colorbar labels
cbar = ax.collections[0].colorbar
cbar.set_ticks([0, 1, 2, 3, 4, 5])
cbar.set_ticklabels(['None', 'Low', 'Moderately low', 'Medium', 'Moderately high', 'High'])  # Centered category labels
cbar.ax.tick_params(labelsize=10)  # Adjust size of colorbar tick labels

# Adjust the position and size of the colorbar
cbar.ax.set_position([0.92, 0.25, 0.03, 0.5])  # Adjust the position to make it smaller

# Remove y-axis ticks and labels
ax.set_yticks([])
ax.set_yticklabels([])

# Save the figure
plt.savefig("plots/042_risk_heatmap_dssaqchis.pdf")

# Show the plot
plt.show()
```
