### Normalization

```python
ds_conn_07["obs_per_origin_area"] = ds_conn_07.obs / ds_conn_07.water_fraction_hex0
```

```python
ds_conn_28["obs_per_origin_area"] = ds_conn_28.obs / ds_conn_28.water_fraction_hex0
```

### Labellig habitable zones
- depth < 90m, rest sites, aquaculture, historic sites, known disease sites, known healthy sites

```python
ds_conn_07["habitable_hex0"] = (
    (ds_conn_07.depth_median_hex0 < 90)
    | (ds_conn_07.his_count_hex0 > 0)
    | (ds_conn_07.rst_count_hex0 > 0)
    | (ds_conn_07.aqc_count_hex0 > 0)
    | (ds_conn_07.dss_count_hex0 > 0)
    | (ds_conn_07.hly_count_hex0 > 0)
).rename("habitable_hex0")

ds_conn_07["habitable_hex1"] = (
    (ds_conn_07.depth_median_hex1 < 90)
    | (ds_conn_07.rst_count_hex1 > 0)
    | (ds_conn_07.his_count_hex1 > 0)
    | (ds_conn_07.aqc_count_hex1 > 0)
    | (ds_conn_07.dss_count_hex1 > 0)
    | (ds_conn_07.hly_count_hex1 > 0)
).rename("habitable_hex1")
```

```python
ds_conn_14["habitable_hex0"] = (
    (ds_conn_14.depth_median_hex0 < 90)
    | (ds_conn_14.his_count_hex0 > 0)
    | (ds_conn_14.rst_count_hex0 > 0)
    | (ds_conn_14.aqc_count_hex0 > 0)
    | (ds_conn_14.dss_count_hex0 > 0)
    | (ds_conn_14.hly_count_hex0 > 0)
).rename("habitable_hex0")

ds_conn_14["habitable_hex1"] = (
    (ds_conn_14.depth_median_hex1 < 90)
    | (ds_conn_14.rst_count_hex1 > 0)
    | (ds_conn_14.his_count_hex1 > 0)
    | (ds_conn_14.aqc_count_hex1 > 0)
    | (ds_conn_14.dss_count_hex1 > 0)
    | (ds_conn_14.hly_count_hex1 > 0)
).rename("habitable_hex1")
```

```python
ds_conn_28["habitable_hex0"] = (
    (ds_conn_28.depth_median_hex0 < 90)
    | (ds_conn_28.his_count_hex0 > 0)
    | (ds_conn_28.rst_count_hex0 > 0)
    | (ds_conn_28.aqc_count_hex0 > 0)
    | (ds_conn_28.dss_count_hex0 > 0)
    | (ds_conn_28.hly_count_hex0 > 0)
).rename("habitable_hex0")

ds_conn_28["habitable_hex1"] = (
    (ds_conn_28.depth_median_hex1 < 90)
    | (ds_conn_28.rst_count_hex1 > 0)
    | (ds_conn_28.his_count_hex1 > 0)
    | (ds_conn_28.aqc_count_hex1 > 0)
    | (ds_conn_28.dss_count_hex1 > 0)
    | (ds_conn_28.hly_count_hex1 > 0)
).rename("habitable_hex1")
```

```python
habitable_to_habitable_01y_07 = ds_conn_07.obs_per_origin_area.where(
    (ds_conn_07.habitable_hex0 > 0)
    & (ds_conn_07.habitable_hex1 > 0)
)
```

```python
habitable_to_habitable_01y_14 = ds_conn_14.obs_per_origin_area.where(
    (ds_conn_14.habitable_hex0 > 0)
    & (ds_conn_14.habitable_hex1 > 0)
)
```

```python
habitable_to_habitable_01y_28 = ds_conn_28.obs_per_origin_area.where(
    (ds_conn_28.habitable_hex0 > 0)
    & (ds_conn_28.habitable_hex1 > 0)
)
```

```python
hex_patches = []
for _lon, _lat in zip(
    ds_conn_07.lon_hex1_corners.data.T,
    ds_conn_07.lat_hex1_corners.data.T,
):
    ll = np.stack((_lon, _lat)).T
    hex_patches.append(Polygon(ll, closed=False))
```

### Diseased to habitable AVERAGE 00-07

```python
dss_to_habitable_01y = ds_conn_07.obs_per_origin_area.where(
    (ds_conn_07.dss_count_hex0 > 0),
    drop=True,
).where(
    (ds_conn_07.depth_median_hex0 < 90)
    | (ds_conn_07.pop_count_hex0 > 0)
    | (ds_conn_07.his_count_hex0 > 0)
    | (ds_conn_07.rst_count_hex0 > 0)
    | (ds_conn_07.aqc_count_hex0 > 0)
    | (ds_conn_07.dss_count_hex0 > 0)
    | (ds_conn_07.hly_count_hex0 > 0)
)

dss_to_habitable_01y
```

```python
habitable = (
    (ds_conn_07.depth_median_hex0 < 90)
    | (ds_conn_07.pop_count_hex0 > 0)
    | (ds_conn_07.his_count_hex0 > 0)
    | (ds_conn_07.rst_count_hex0 > 0)
    | (ds_conn_07.aqc_count_hex0 > 0)
    | (ds_conn_07.dss_count_hex0 > 0)
    | (ds_conn_07.hly_count_hex0 > 0)
)
```

```python
dss_origins_01y = dss_to_habitable_01y.sum(["month", "year", "hex1"]).compute().to_dataset().to_pandas()
dss_origins_01y.where(dss_origins_01y["obs_per_origin_area"] > 0).dropna()
```

```python
dss_destinations_01y = dss_to_habitable_01y.sum(["month", "year", "hex0"]).compute().to_dataset().to_pandas()
dss_destinations_01y
```

```python
dss_origin_count_01y = (dss_to_habitable_01y.sum(["month", "year"]) > 0).sum("hex0").compute().to_dataset().to_pandas()
dss_origin_count_01y
```

```python
dss_to_habitable_stats = (
    dss_origin_count_01y
    .where(dss_origin_count_01y["obs_per_origin_area"] > 0).dropna()
    .sort_values("obs_per_origin_area", ascending=False)
    .rename(columns={"lon_hex1": "lon", "lat_hex1": "lat", "obs_per_origin_area": "num_dss_sources"})
)

dss_to_habitable_stats = dss_to_habitable_stats.join(
    dss_destinations_01y
    .where(dss_destinations_01y["obs_per_origin_area"] > 0).dropna()
    .sort_values("obs_per_origin_area", ascending=False)
    .rename(columns={"obs_per_origin_area": "rel_pathogen_concentration"})
    [["rel_pathogen_concentration"]]
)

dss_to_habitable_stats["rel_pathogen_concentration"] /= dss_to_habitable_stats["rel_pathogen_concentration"].max()

dss_to_habitable_stats["rel_num_dss_sources"] = dss_to_habitable_stats["num_dss_sources"] / dss_to_habitable_stats["num_dss_sources"].max()

dss_to_habitable_stats.where(dss_to_habitable_stats.rel_pathogen_concentration > 0).dropna()
```

```python
dss_to_habitable_stats
```

```python
dss_to_habitable_stats["rel_num_dss_sources_category"] = pd.cut(
    dss_to_habitable_stats.rel_num_dss_sources,
    bins=[0, 1/5, 2/5, 3/5, 4/5, 1],
    labels=[1, 2, 3, 4, 5],
).astype(float)
dss_to_habitable_stats["rel_pathogen_concentration_category"] = pd.cut(
    dss_to_habitable_stats.rel_pathogen_concentration,
    bins=[0, 0.01, 0.05, 0.1, 0.25, 1],
    labels=[1, 2, 3, 4, 5],
).astype(float)
dss_to_habitable_stats
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from matplotlib.collections import PatchCollection
from matplotlib.colors import ListedColormap, BoundaryNorm
import numpy as np
import pandas as pd
from shapely.geometry import Polygon

# Create a mapping of hex IDs to patches
hex_to_patch = {
    hex_id: patch for hex_id, patch in zip(ds_conn_07.hex1.values, hex_patches)
}

# Extract category data and ensure correct order
hex_ids = dss_to_habitable_stats.index
category_data = dss_to_habitable_stats["rel_pathogen_concentration_category"]

# Prepare the patches and data for plotting
hex_patches_filtered = []
category_values = []

for hex_id in hex_ids:
    if hex_id in hex_to_patch:
        hex_patches_filtered.append(hex_to_patch[hex_id])
        category_values.append(category_data.loc[hex_id])

# Define categorical colors for each category
categories = [1, 2, 3, 4, 5]
category_colors = ['#c8e6c9', '#e6ee9c' ,'#fff176','#ff8a65','#f44336']  # Light green, light yellow, light orange, red
cmap = ListedColormap(category_colors)
norm = BoundaryNorm(boundaries=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5], ncolors=len(categories))

# Create the plot
fig, ax = plt.subplots(
    1, 1,
    subplot_kw={"projection": ccrs.Mercator()},
    figsize=(12, 8),
)

# Create the PatchCollection
p = PatchCollection(
    hex_patches_filtered,
    cmap=cmap,
    norm=norm,
    alpha=1.0,  
    linewidth=0.1,
    transform=ccrs.PlateCarree(),
    zorder=0
)

# Set the data array for coloring the patches
p.set_array(np.array(category_values))

# Add the collection to the plot
ax.add_collection(p, autolim=True)

# Extract category data and ensure correct order
hex_ids = dss_to_habitable_stats.index
category_data = dss_to_habitable_stats["rel_pathogen_concentration_category"]

# Prepare the patches and data for plotting
hex_patches_filtered = []
category_values = []

for hex_id in hex_ids:
    if hex_id in hex_to_patch:
        hex_patches_filtered.append(hex_to_patch[hex_id])
        category_values.append(category_data.loc[hex_id])

# Find the intersection of hex IDs that are plotted and that overlap with land
#plot_hex_ids_with_land_overlap = set(hex_ids).intersection(land_overlap_hex_ids)

# Select patches corresponding to hexes with land overlap
#land_overlap_hexes = [hex_to_patch[hex_id] for hex_id in plot_hex_ids_with_land_overlap if hex_id in hex_to_patch]

#land_overlap = PatchCollection(
#    land_overlap_hexes,
#    color='grey',
#    alpha=1,
#    linewidth=0.1,
#    transform=ccrs.PlateCarree(),
#    zorder=2
#)

#ax.add_collection(land_overlap, autolim=True)

# Set the extent of the plot (longitude and latitude limits)
ax.set_extent([-12, 9, 46, 62])

# Add coastlines and gridlines
ax.coastlines(zorder=0)
gl = ax.gridlines(draw_labels=True, zorder=0)
gl.xlabel_style = {'size': 16}
gl.ylabel_style = {'size': 16}
gl.top_labels = False   # suppress top labels
gl.right_labels = False # suppress right labels

# Set the title of the plot
#ax.set_title("Pathogen Concentration Categories", fontsize=16)

# Add colorbar with custom formatting
cbar = fig.colorbar(p, ax=ax, orientation='vertical', pad=0.02, shrink=0.5, aspect=15)
cbar.ax.tick_params(labelsize=16)
cbar.set_ticks(categories)  # Set colorbar ticks to match your categories
cbar.set_ticklabels(['Very low', 'Low', 'Moderate', 'Moderately high', 'High'])  # Label each tick with your category values

# Format colorbar to be more condensed
cbar.ax.invert_yaxis()  # Optional: invert the colorbar to align with the plot orientation

# Save the figure
fig.savefig("plots/042_risk-analysis_dss_15m_00-07.pdf")
```
